# creating a metadata for the preprocess function


```python
df = pd.DataFrame(
    mapping,
    columns=[
        "cell_name",
        "self_reported_ethnicity_ontology_term_id",
        "tissue_ontology_term_id",
        "disease_ontology_term_id",
        "sex_ontology_term_id",
        "development_stage_ontology_term_id",
    ],
)
df["self_reported_ethnicity_ontology_term_id"] = df[
    "self_reported_ethnicity_ontology_term_id"
].str.lower()
df["tissue_ontology_term_id"] = df["tissue_ontology_term_id"].str.lower()
df["disease_ontology_term_id"] = df["disease_ontology_term_id"].str.lower()
df["sex_ontology_term_id"] = df["sex_ontology_term_id"].str.lower()
df = df.replace(convert)
for i in range(len(df)):
    if df["development_stage_ontology_term_id"][i] == "unknown":
        age_group = "unknown"
    elif df["development_stage_ontology_term_id"][i] == "adult":
        age_group = "HsapDv:0000258"
    elif df["development_stage_ontology_term_id"][i] > 80:
        age_group = "HsapDv:0000095"
    elif df["development_stage_ontology_term_id"][i] > 60:
        age_group = "HsapDv:0000272"
    elif df["development_stage_ontology_term_id"][i] > 40:
        age_group = "HsapDv:0000258"
    elif df["development_stage_ontology_term_id"][i] > 20:
        age_group = "HsapDv:0000266"
    elif df["development_stage_ontology_term_id"][i] < 4:
        age_group = "HsapDv:0000265"
    else:
        age_group = "HsapDv:0000271"
    df["development_stage_ontology_term_id"][i] = age_group
df["assay_ontology_term_id"] = "EFO:0009922"
df["cell_type_ontology_term_id"] = "unknown"
```

```python
def additional_preprocess(adata):
    # Map df columns to adata.obs using cell_name as the key
    for col in df.columns[1:]:
        adata.obs[col] = adata.obs["cell_name"].map(df.set_index("cell_name")[col])
    adata.obs["cell_culture"] = True
    adata.obs["organism_ontology_term_id"] = "NCBITaxon:9606"
    adata.obs["is_primary_data"] = True
    adata.obs["cell_type"] = "unknown"
    return adata
```
